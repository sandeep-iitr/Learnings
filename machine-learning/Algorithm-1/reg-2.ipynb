```python
import pandas as pd
import numpy as np
headers = ["", "a", "b", "c", "d",
           "e", "f", "g", "h",
           "y"]
df = pd.read_csv('intern_data.csv',na_values="?")

print('-'*100)
print('Data visualization')
print('-'*100)
print(df.head())
print('-'*100)
print(df.dtypes)
```

# Changing Objects to Category
- Used in one hot encoding

```python
df["c"] = df["c"].astype('category')
df["h"] = df["h"].astype('category')

# Renaming the Unnamed column
df.rename(columns={'Unnamed: 0':'ID'}, inplace=True)

df.dtypes
```

# One Hot encoding

```python
df=pd.get_dummies(df, columns=["c"])
df=pd.get_dummies(df, columns=["h"])
print(df.head())
```

## This displays a table of detailed distribution information for each of the attributes in our data 

```python
print(df.describe())
```

```python
cdf=df.drop(labels='ID',axis=1)
cdf.head()
```
