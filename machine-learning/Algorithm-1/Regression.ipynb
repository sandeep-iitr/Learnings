# Load the Dataset

```python
import csv
data=[]
labels=[]
with open('intern_data.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        #print(row['a'], row['b'],row['c'],row['d'])
        
        data.append([float(row['a']), float(row['b']),row['c'],float(row['d']),float(row['e']),float(row['f']),float(row['g']),row['h']])
        labels.append(float(row['y']))
#print(data)
#print(labels)
```

```python
import pandas as pd
import numpy as np
headers = ["", "a", "b", "c", "d",
           "e", "f", "g", "h",
           "y"]
df = pd.read_csv('intern_data.csv',na_values="?")

print('-'*100)
print('Data visualization')
print('-'*100)
print(df.head())
print('-'*100)
print(df.dtypes)

obj_df = df.select_dtypes(include=['object']).copy()
print('-'*100)
print(obj_df.head())
```

```python
obj_df["c"].value_counts()
```

```python
obj_df["h"].value_counts()
```

```python
obj_df["c"] = obj_df["c"].astype('category')
obj_df["h"] = obj_df["h"].astype('category')

obj_df.dtypes
```

```python
df["c"] = df["c"].astype('category')
df["h"] = df["h"].astype('category')

# Renaming the Unnamed column
df.rename(columns={'Unnamed: 0':'ID'}, inplace=True)

df.dtypes
```

# One hot encoding

```python
df=pd.get_dummies(df, columns=["c"])
df=pd.get_dummies(df, columns=["h"])
print(df.head())
```

```python
from IPython.display import HTML, display
import statsmodels.api as sm
from statsmodels.formula.api import ols

housing_model = ols("y ~ a+b+d+e+f+g+y", data=df).fit()
# summarize our model
housing_model_summary = housing_model.summary()

```

```python
HTML(
housing_model_summary\
.as_html()\
.replace(' Adj. R-squared: ', ' Adj. R-squared: ')\
.replace('coef', 'coef')\
.replace('std err', 'std err')\
.replace('P>|t|', 'P>|t|')\
.replace('[95.0% Conf. Int.]', '[95.0% Conf. Int.]')
)
```

```python
# this produces our four regression plots for total_unemployed

# plots the graphs inside the Jupyter/IPython Notebook
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt

# use figsize to adjust size, and pass in to plot_regress_exog
fig = plt.figure(figsize=(15,8))

# pass in the model as the first parameter, then specify the 
# predictor variable we want to analyze
fig = sm.graphics.plot_regress_exog(housing_model, "b", fig=fig)
```
